# **Exploring Polynomial Regression for Predicting Position Salaries**

This notebook explores polynomial regression for salary prediction based on position levels. It trains and visualizes both linear and polynomial regression models with degrees 2 and 4, and predicts salaries for a new position level.

# **Polynomial Regression**photo_2024-08-27_16-06-41.jpg

**Polynomial Regression** is an extension of linear regression that models the relationship between the independent variable **X** and the dependent variable
**y** as an n-th degree polynomial. Unlike linear regression, which fits a straight line to the data, polynomial regression fits a curve, allowing it to capture more complex relationships.

## Importing the libraries

```
# import
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

This dataset contains salary information for various job positions at different levels within an organization. Each entry includes:

* **Position**: The job title or role within the company.
* **Level**: A numerical level indicating the seniority or rank of the position.
* **Salary**: The corresponding annual salary for the position, provided in USD.

```
# Read datset
dataset = pd.read_csv('Position_Salaries.csv')
```

```
# Print the dataset
print(dataset)
```

```
# Devide to dependent and the Independent variable arrays
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```
# Print the independent variable array
print(X)
```

```
# Print the dependent variables array
print(y)
```

## Training the Linear Regression model on the whole dataset

```
# Implement the Linear regression model
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()

# fit the model with X and y
lin_reg.fit(X, y)
```

## Visualising the Linear Regression results

```
# Plot the actual data points (real salary) and the predicted points of liner regression model
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg.predict(X), color = 'blue')
plt.title('Linear Regression results')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

## Training the Polynomial Regression model on the whole dataset with degree = 2

```
from sklearn.preprocessing import PolynomialFeatures

# Implement the PolinomialRegression model
poly_reg = PolynomialFeatures(degree = 2)   # ( bo + b1X1 + b2X2 )
X_poly = poly_reg.fit_transform(X)

# fit x poly with linear regression model
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

## Visualising the Polynomial Regression results

```
# Plot the actual data points and the predicted points of Polinomial regression model
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)), color = 'blue')
plt.title('Polynomial Regression results')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

## Visualising the Polynomial Regression results with smoother curve

```
# Plot the actual data points and the predicted points of the Polynomial regression model with smoother curve
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, lin_reg_2.predict(poly_reg.fit_transform(X_grid)), color = 'blue')
plt.title(' Smoothed Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

## Training the Polynomial Regression model on the whole dataset with degree = 4

```
from sklearn.preprocessing import PolynomialFeatures

# Implement the PolinomialRegression model
poly_reg = PolynomialFeatures(degree = 4)   # ( bo + b1X1 + b2X2 + b3X3 + b4X4 )
X_poly = poly_reg.fit_transform(X)

# fit x poly with linear regression model
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

## Visualising the Polynomial Regression results

```
# Plot the actual data points and the predicted points of Polinomial regression model
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)), color = 'blue')
plt.title('Polynomial Regression results')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

## Visualising the Polynomial Regression results with smoother curve

```
# Plot the actual data points and the predicted points of the Polinomial regression model with smoother curve
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, lin_reg_2.predict(poly_reg.fit_transform(X_grid)), color = 'blue')
plt.title(' Smoothed Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

## Predicting a new result with Linear Regression

```
# get a prediction for new level 6.5 using the regression model

new_level = 6.5
lin_reg.predict([[new_level]])
```
